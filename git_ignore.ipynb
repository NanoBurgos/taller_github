# **.gitignore**

En Git, .gitignore es un archivo de configuración que indica qué archivos o carpetas NO deben ser rastreados ni versionados por el repositorio.

Es fundamental para evitar subir archivos locales, temporales o sensibles.

# **¿Para qué sirve .gitignore?**

Se utiliza para excluir, por ejemplo:

- Entornos virtuales (.venv/, env/)

- Dependencias (node_modules/)

- Archivos compilados (.exe, .class, .pyc)

- Configuraciones locales (.env)

- Archivos del sistema (.DS_Store, Thumbs.db)

- Archivos generados automáticamente (logs, cachés)

# **¿Cómo funciona?**

Git lee el archivo .gitignore en la raíz del repositorio y aplica las reglas en orden.

# **Reglas comunes de .gitignore**

|Regla	|Significado|
|:-:|:-:|
|archivo.txt	|Ignora un archivo específico|
|carpeta/	|Ignora una carpeta|
|*.log	|Ignora todos los archivos .log|
|!importante.log	|Excepción (NO se ignora)|
|/config.json	|Solo en la raíz|
|**/temp	|En cualquier nivel|

# **Punto clave (error muy común)**

**.gitignore** NO elimina archivos que ya están en el repositorio

Si un archivo ya fue agregado con git add, Git lo seguirá rastreando aunque lo pongas en .gitignore.

Solución:

        git rm --cached archivo.txt

        git commit -m "Eliminar archivo del tracking"

# **¿Dónde crear el .gitignore?**

- En la raíz del repositorio

- Nombre exacto: .gitignore (con punto adelante)

- Se versiona como cualquier otro archivo

# **Relación con GitHub**

- GitHub no aplica .gitignore automáticamente

- Al crear un repo nuevo, GitHub puede sugerirte uno base (Python, Node, etc.)

- Puedes combinar varios .gitignore (por lenguaje)
