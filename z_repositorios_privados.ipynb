**Si activas:**

            “Include private contributions on my profile”

**Entonces:**

- Tus commits en repos privados sí suman al gráfico de contribuciones

- No se muestra el contenido, repositorio ni detalles

- Solo aparece como contribuciones anónimas (sin revelar info sensible)

## **🔒 Si es un repositorio privado**

**En tu perfil de GitHub:**

- NO se mostrará el mensaje del commit
- NO se verá el nombre del repositorio
- NO se podrá acceder al código

## **🌐 Si es un repositorio público**

**Ahí sí se ve todo:**

- Mensaje del commit
- Nombre del repositorio
- Archivos modificados
- Historial completo

## **🧠 Importante entender**

**GitHub separa dos cosas:**

- Tu actividad (contribution graph) → sí incluye privados (si activaste la opción)
- El detalle visible → solo muestra lo público

**Ejemplo claro**

Supongamos que haces esto:

- 5 commits en un repo privado
- 2 commits en uno público

**En tu perfil se verá:**

- 7 contribuciones ese día
- Pero solo podrás abrir y ver los 2 commits públicos

## **⚠️ Excepción**

Si alguien tiene acceso a tu repo privado (por ejemplo, en una organización o colaborador):

- Sí verá el historial completo de commits dentro de ese repo

## **Todo el repositorio hereda la misma visibilidad en GitHub.**

**🔒 Caso: repositorio privado**

Si creas un repo como privado:

✅ Todas las ramas son privadas (main, dev, feature/*, etc.)
✅ Todo el historial de commits es privado
✅ Issues, PRs, releases, TODO queda privado

**No existe un caso donde:**

            “main es privado pero otras ramas son públicas”

Eso no es posible dentro de un mismo repo.

## **🧠 Cómo funciona realmente**

La visibilidad se define a nivel de:

- Repositorio completo, no por rama

Las ramas son solo punteros a commits dentro de ese repo, no entidades independientes en cuanto a privacidad.

## **🔄 ¿Cómo hacer algo “semi-público” entonces?**

Si quieres mostrar algo pero mantener otras partes privadas, tienes que:

- Crear otro repositorio público
- Subir ahí solo lo que quieres exponer (por ejemplo una versión limpia)

**O usar:**

- forks públicos (con cuidado)
- o repos separados (project-core privado + project-demo público)

## **⚠️ Detalle importante**

**Si haces el repo público después:**

- Todas las ramas pasan a ser públicas automáticamente
- Todo el historial también
