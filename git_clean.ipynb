# **git clean**

git clean elimina archivos que Git NO está siguiendo (untracked files) del working directory.

No toca:

- commits,

- staging,

- archivos trackeados.

Es una herramienta de limpieza, no de control de historial.


# **Conceptos clave antes de usarlo**

Tipos de archivos

|Tipo	|Estado|
|:-:|:-:|
|Trackeados	|ya están en Git|
|Modificados	|cambios locales|
|Untracked	|nunca se agregaron|
|Ignorados	|listados en .gitignore|

git clean actúa solo sobre untracked

(y opcionalmente ignorados).

# **Regla de oro (muy importante)**

Nunca ejecutes git clean sin antes simularlo.

# **Simulación (modo seguro)**

            git clean -n


Salida típica:

            Would remove build/

            Would remove temp.log


Nada se borra. Solo te muestra qué pasaría.

# **Limpieza real de archivos**

            git clean -f


Qué hace:

- borra archivos no trackeados,

- no borra carpetas.

# **Borrar carpetas no trackeadas**

            git clean -fd


Caso real:

- eliminar carpetas build/, dist/, __pycache__/.

# **Incluir archivos ignorados**

            git clean -fx


**Esto borra archivos listados en .gitignore.**

Caso real:

- limpiar completamente un entorno de build.

# **Combinación recomendada (segura)**

            git clean -nfd

            git clean -fd


Primero mirás, luego ejecutás.

# **Limpiar solo ciertas rutas**

            git clean -fd logs/


Caso real:

- limpiar solo una carpeta generada.

# **Diferencia con git reset y git restore**

|Comando	|Borra	|Toca commits|
|:-:|:-:|:-:|
|clean	|archivos no trackeados	|❌|
|restore	|cambios en archivos	|❌|
|reset	|staging / commits	|✔️|

# **Flujo profesional típico**

Proyecto sucio luego de pruebas

            git status

            git clean -nfd

            git clean -fd

Antes de un build o deploy

            git clean -nfx

            git clean -fx

# **Errores comunes**

Error crítico

            git clean -fx


sin revisar.

**Consecuencia:**

- pérdida irreversible de archivos generados.

# **Buena práctica**

Usar siempre:

            git status

            git clean -n

# **Regla mental clave**

**git clean** borra archivos que Git nunca cuidó.

Si te importa, debería estar commiteado o stasheado.

# **Resumen rápido**

git clean -n _______________ simulación

git clean -f _______________ borra archivos

git clean -fd _______________ incluye carpetas

git clean -fx _______________ incluye ignorados

git clean -nfd _______________ combinación segura
