## **Etiquetas - Tags**

Las etiquetas **(tags)** en Git son una forma de marcar puntos importantes en el historial del proyecto. Se usan muchísimo para identificar versiones, como:

- v1.0.0
- v1.1.0
- v2.0.0

Cuando subís esos tags a GitHub, aparecen en la sección de Releases y Tags del repositorio.

### **2. Ver etiquetas existentes**

```bash 
git tag
```

Salida:

- v1.0.0
- v1.1.0
- v2.0.0

### **3. Crear una etiqueta simple**

```bash
git tag v1.0.0
```

Esto crea un tag en el commit actual (HEAD).

### **4. Crear una etiqueta en un commit específico**

Primero obtenés el hash:

```bash
git log --oneline
```

**Ejemplo:**

a3f1c92 Corrige bug

b81e2ad Agrega login

c0921fa Proyecto inicial

Luego:

```bash
git tag v1.0.0 b81e2ad
```

Ahora **v1.0.0** apunta al commit **b81e2ad.**

### **5. Tags anotados (los más usados)**

En proyectos profesionales se prefieren los annotated tags.

```bash
git tag -a v1.0.0 -m "Primera versión estable"
```

**Esto guarda:**

- nombre,
- autor,
- fecha,
- descripción.

### **6. Ver información de un tag**

```bash
git show v1.0.0
```

Muestra:

- el commit asociado,
- el mensaje del tag,
- los cambios del commit.

### **7. Subir tags a GitHub**


Los tags no se envían automáticamente con git push.

Subir un tag
```bash
git push origin v1.0.0
```

Subir todos los tags
```bash
git push origin --tags
```

### **8. Descargar tags de GitHub**

```bash
git fetch --tags```


### **9. Cambiarte a una versión etiquetada**
```bash
git checkout v1.0.0
```

o usando comandos modernos:

```bash
git switch --detach v1.0.0
```

Esto te permite inspeccionar esa versión.

- Quedás en estado DETACHED HEAD.

### **10. Eliminar tags**

Localmente
```bash
git tag -d v1.0.0
```

En GitHub
```bash
git push origin --delete v1.0.0
```

### **11. Convención de versionado**

### La más utilizada es Semantic Versioning (SemVer).

Formato:

MAJOR.MINOR.PATCH

Ejemplo:

1.0.0

1.1.0

1.1.1

2.0.0

**PATCH**

Corrección de errores.

1.0.0 → 1.0.1

**MINOR**

Nueva funcionalidad compatible.

1.0.0 → 1.1.0

**MAJOR**

Cambios incompatibles.

1.0.0 → 2.0.0

El estándar está definido por el concepto de **Semantic Versioning.**

### **12. Caso real para tus repositorios**

Supongamos que terminás tu proyecto.

Podrías hacer:

```bash
git add .
git commit -m "Versión estable del sistema"
git tag -a v1.0.0 -m "Primera versión funcional"
git push
git push origin v1.0.0
```

Más adelante:
```bash
git tag -a v1.1.0 -m "Panel web mejorado"
git push origin v1.1.0
```
Así cualquier persona podrá descargar exactamente la versión que corresponda.

### **13. Diferencia entre rama y etiqueta**

|Rama	|Tag|
|:-:|:-:|
|Se mueve con nuevos commits	|Queda fija|
|Desarrollo activo	|Punto histórico|
|main, develop, feature-x	|v1.0.0, v2.0.0|
|Cambia constantemente	|Nunca cambia|

**La regla mental más útil es:**

- Las ramas representan trabajo en progreso. 
- Los tags representan versiones congeladas del proyecto.

Por eso, cuando publiques proyectos en GitHub, los tags son la forma estándar de marcar entregas, hitos importantes y versiones estables.
