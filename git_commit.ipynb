# **git commit**

guarda una “foto” (snapshot) del estado actual del staging area (índice) en el historial del repositorio.

**En términos simples:**

git add → selecciona qué cambios van a entrar

git commit → registra esos cambios de forma permanente en la rama actual

**Cada commit tiene:**

- Un hash (identificador único)

- Un mensaje

- Autor y fecha

- Referencia al commit anterior

# **Uso básico**

            git commit


Abre el editor configurado (vim, nano, etc.)

Permite escribir un mensaje de commit

Ejemplo de mensaje recomendado:

Agrega validación de datos en formulario


# **git commit -m (mensaje directo)**

            git commit -m "Corrige error en cálculo de impuestos"


- Variante más común

- Evita abrir el editor

- Ideal para commits simples y claros

**Buena práctica:** mensajes cortos, en presente y descriptivos.

# **git commit -a (add + commit)**

            git commit -a -m "Actualiza README"


**Hace automáticamente:**

- git add solo de archivos ya rastreados

- luego el commit

**Importante:**

- NO incluye archivos nuevos (sin seguimiento)

- No reemplaza completamente a git add

# **git commit --amend (modificar el último commit)**
            
            git commit --amend


**Usos típicos:**

- Corregir el mensaje del último commit

- Agregar archivos que olvidaste incluir

**Ejemplo:**

            git add archivo_olvidado.py

            git commit --amend


**Advertencia:**

- Reescribe el historial

- No usar si el commit ya fue pusheado a un repositorio compartido

# **git commit --amend -m**

            git commit --amend -m "Mensaje corregido"


- Cambia el mensaje sin abrir el editor

# **git commit --no-edit**

            git commit --amend --no-edit


- Mantiene el mensaje anterior

- Útil cuando solo agregas archivos faltantes

# **git commit --dry-run**

            git commit --dry-run


- Simula el commit

- Muestra qué archivos entrarían

- No guarda nada

Útil para verificar antes de confirmar cambios.

# **git commit -v (verbose)**

            git commit -v


- Muestra el diff de los cambios en el editor

- Ayuda a revisar antes de confirmar

# **git commit --allow-empty**

            git commit --allow-empty -m "Inicializa flujo del proyecto"


- Crea un commit sin cambios

- Útil para:

Marcar hitos

Disparar pipelines CI/CD

# **Flujo típico recomendado**
            
            git status
            
            git add archivo.py
            
            git commit -m "Implementa lógica de autenticación"
            
            git push

# **Buenas prácticas de commits**

- Un commit = una idea o cambio lógico

- Mensajes claros y específicos

- Evitar commits gigantes

- No usar --amend en commits ya compartidos

- Preferir commits pequeños y frecuentes
