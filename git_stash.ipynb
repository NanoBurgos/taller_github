# **git stash **

git stash guarda temporalmente cambios no commiteados (working directory y, opcionalmente, staging) y deja tu árbol de trabajo limpio, como si recién hubieras hecho checkout a la rama.

Es ideal cuando:

- tenés trabajo a medio hacer,

- necesitás cambiar de rama urgente,

- no querés hacer un commit incompleto.

# **Qué guarda y qué NO guarda**

Guarda por defecto

- archivos modificados trackeados,

- cambios en el working directory,

- cambios en staging.

NO guarda

- archivos no trackeados (por defecto),

- archivos ignorados (.gitignore).

# **Uso básico**

Guardar cambios

            git stash


Resultado:

- tu working directory queda limpio,

- los cambios se guardan en una pila (stack).

# ** Ver stashes guardados

            git stash list


Salida típica:

            stash@{0}: WIP on main: 3a1c92 Agrega validación

            stash@{1}: WIP on feature-login: f91e3aa Ajusta UI

# **Recuperar el último stash**

            git stash pop

Qué hace:

- aplica los cambios,

- elimina ese stash de la lista.

Aplicar sin borrar

            git stash apply


Útil si:

- querés probar algo,

- no estás seguro de descartar el stash.

# **Stash con mensaje (muy recomendado)**

            git stash push -m "WIP login validation"


Buena práctica:

- siempre usar mensajes descriptivos.

Incluir archivos no trackeados

            git stash -u


O:

            git stash push -u -m "Incluye archivos nuevos"


Caso real:

- creaste archivos nuevos que todavía no querés commitear.

# **Stash selectivo (por archivo)//

            git stash push archivo.py


Caso real:

- querés stashear solo parte del trabajo.

# **Aplicar un stash específico**

            git stash apply stash@{1}


O:

            git stash pop stash@{1}

# **Eliminar stashes**

Borrar uno puntual

            git stash drop stash@{0}

Borrar todos
            
            git stash clear


Irreversible.

# **Stash y conflictos**

Si al aplicar un stash hay conflictos:

- Git te lo avisa,

- los resolvés como un merge normal,

- luego hacés commit.

# **Flujo profesional típico**

Escenario real

Estás trabajando en feature-x

aparece un bug urgente en main

            git stash push -m "WIP feature-x"
            
            git switch main
            
            git pull
            
            # arreglás el bug
            
            git commit -m "Hotfix crítico"
            
            git switch feature-x
            
            git stash pop

# **Errores comunes**

**Error: usar stash como commits**

+ mal hábito.

stash es temporal, no historial.

**Error: olvidar stashes**

Solución:

            git stash list


revisarlo periódicamente.

# **Regla mental clave**

**git stash** es una pausa, no un guardado definitivo.

Si algo es importante, debe terminar en un commit.
