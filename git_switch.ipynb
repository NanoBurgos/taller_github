# **git switch**

git switch se usa exclusivamente para trabajar con ramas:

- cambiar de rama

- crear ramas nuevas

- moverse entre ramas locales y remotas

No restaura archivos. No modifica commits.

Fue creado para **evitar errores** comunes de git checkout.

# **Cambio básico de rama**

            git switch main


Qué sucede:

- HEAD apunta a main

- el working directory se actualiza

Caso real:

- volver a main para hacer pull o revisar estado estable.

# **Crear y cambiar a una rama nueva**

            git switch -c feature-login


Equivale a:

            git branch feature-login
          
            git switch feature-login


Caso real:

- empezar una nueva funcionalidad o experimento.

# **Volver a la rama anterior (muy útil)**

            git switch -


Caso real:

- alternar rápido entre dos ramas.

# **Cambiar a una rama remota**

Primero asegurate de tenerla:

            git fetch


Luego:

            git switch feature-api


Si no existe localmente:

            git switch -c feature-api origin/feature-api


Caso real:

- empezar a trabajar sobre una rama creada por otra persona.

# **Forzar cambio de rama (con cuidado)**

            git switch -f main


Qué hace:

- descarta cambios locales que impiden el cambio.

Usar solo si sabés que no necesitás esos cambios.

# **Estados inválidos que evita git switch**

Con checkout era común:

- quedar en DETACHED HEAD sin saberlo,

- mezclar cambio de ramas con restauración de archivos.

Con git switch:

- no podés restaurar archivos,

- no podés hacer checkout a un commit sin rama (por defecto).

Esto es intencional.

# **git switch vs git checkout**

|Acción	|checkout	|switch|
|:-:|:-:|:-:|
|Cambiar rama	|✔️|	✔️|
|Crear rama	|✔️	|✔️|
|Restaurar archivos	|✔️	|❌|
|Cambiar a commit	|✔️	|❌|
|Seguridad	|Media	|Alta|

# **Flujo profesional recomendado**

            git status

            git switch main

            git pull

            git switch -c feature-x


Y al terminar:

            git switch main

            git merge feature-x

# **Errores comunes**

Error: “no puedo cambiar de rama”

Motivo:

- archivos modificados sin commit.

Solución:

            git stash

            git switch main

            git stash pop

# **Casos de uso reales**

Alternar rápidamente entre tareas

            git switch -

Crear rama para hotfix

            git switch -c hotfix-login

Trabajar con ramas remotas

            git fetch

            git switch -c fix origin/fix

# **Regla mental clave**

git switch es solo para ramas.

- Si pensás en archivos → git restore.

- Si pensás en commits → git log / git show.
