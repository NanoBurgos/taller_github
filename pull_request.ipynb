# **Pull Request (PR)**

Es una solicitud para que los cambios que hiciste en una rama (branch) de un repositorio sean integrados (mergeados) en otra rama, normalmente la rama principal (main/master).

# **Analogía simple:**

Imagina que estás editando un documento colaborativo (como Google Docs):

1. Haces una copia del documento original

2. Realizas tus cambios en esa copia

3. Solicitas revisión a tus colegas antes de que tus cambios se incluyan en el documento original

4. Ellos revisan, comentan, sugieren mejoras

5. Cuando está aprobado, tus cambios se fusionan con el original

# **Flujo típico de un Pull Request:**

1. **Creas una rama nueva desde la rama principal**


            git checkout -b mi-nueva-funcionalidad

2. **Trabajas en esa rama - agregas características, corriges bugs, etc.**


            git add .

            git commit -m "Agrego nueva funcionalidad"
            
            git push origin mi-nueva-funcionalidad

3. **Creas el PR en GitHub/GitLab:**

- Vas al repositorio en la plataforma

- Creas una nueva "Pull Request"

- Seleccionas: rama origen (tus cambios) → rama destino (main/master)

4. **Revisión y discusión:**

- Otros desarrolladores revisan tu código

- Hacen comentarios, sugieren cambios

- Pueden aprobar o solicitar modificaciones

5. **Integración:**

- Una vez aprobado, se fusiona (merge) con la rama principal

- La rama temporal se puede eliminar

# **¿Por qué son importantes los PR?**

✅ Revisión de código - Otros ven tu trabajo antes de integrarlo

✅ Control de calidad - Se detectan errores temprano

✅ Conocimiento compartido - El equipo aprende del código de otros

✅ Documentación - Los PRs quedan como registro histórico

✅ Integración continua - Se pueden ejecutar tests automáticamente

✅ Discusión técnica - Espacio para debatir soluciones

# **Partes de un buen PR:**

- Título claro - "Agrega autenticación con Google"

- Descripción detallada - Qué hace, por qué, cómo

- Reviewers - Personas que deben revisarlo

- Labels - etiquetas como "bug", "feature", "documentation"

- Linked issues - relacionado con tickets/tareas

Los Pull Requests son fundamentalmente una herramienta de colaboración y garantía de calidad en el desarrollo de software moderno
