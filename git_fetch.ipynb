# **GIT FETCH**

En Git, fetch es la operación que consulta el repositorio remoto y descarga la información más reciente, sin modificar tu código local ni tu rama actual.

**Definición clara**

git fetch:

- Se conecta al remoto (por ejemplo, origin).

- Descarga commits nuevos, ramas y etiquetas.

- Actualiza las referencias remotas locales (como origin/main).

- **No hace merge ni cambia archivos de tu working directory.**

Es, en esencia, una acción segura y no intrusiva para ver qué cambió en el remoto.

# **Qué NO hace fetch**

- No mezcla cambios con tu rama (merge).

- No modifica archivos locales.

- No crea commits nuevos en tu rama actual.

# **Ejemplo práctico**

Supongamos que trabajás en main y alguien actualiza el repositorio en GitHub.

1. Ejecutás:

            git fetch origin


2. Git descarga los cambios y ahora:

- Tu rama local: main

- La referencia remota: origin/main (más adelantada)

3. Podés comparar:

            git log main..origin/main


Esto te muestra qué commits existen en el remoto pero no en tu local.

# **Flujo correcto de actualización (recomendado)**

1. Ver estado actual:

            git status


2. Traer cambios del remoto:

            git fetch origin


3. Revisar diferencias:

            git diff main origin/main


4. Integrar cambios (si estás de acuerdo):

            git merge origin/main

# **Diferencia clave: fetch vs pull**
|Comando	|Qué hace|
| :-: | :-: |
|git fetch	|Descarga cambios sin aplicarlos|
|git pull	|fetch + merge automático|

**Por eso, cuando querés control total y evitar merges inesperados, se recomienda:**

            git fetch + git merge


en lugar de **git pull** directo.

**git fetch** sirve para “mirar qué pasó en el remoto sin tocar tu código”.
